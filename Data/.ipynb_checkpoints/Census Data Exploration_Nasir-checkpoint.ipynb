```python
import geopandas as gpd
```

```python
gdf = gpd.read_file('Census_Tracts_2020-Copy1.geojson')
```

```python
gdf.head()
```

```python
gdf['CT20'] = gdf['CT20'].astype(str)
df['tract'] = df['tract'].astype(str)
```

```python
merged_gdf = gdf.merge(df, left_on='CT20', right_on='tract')
```

```python
import folium
```

```python
import matplotlib.pyplot as plt
```

```python
print(merged_gdf.is_valid)
```

```python
merged_gdf = merged_gdf[merged_gdf.is_valid]
```

```python
invalid_geom = merged_gdf.is_valid
if invalid_geom.any():
    print(f"Invalid geometries detected: {invalid_geom.sum()}")
```

```python
merged_gdf = merged_gdf[invalid_geom]
```

```python
merged_gdf.reset_index(inplace=True)
```

```python
df['tract_six_digits'] = df['tract'].astype(str).str[-6:]
```

```python
merged_gdf = gdf.merge(df, left_on='CT20', right_on='tract_six_digits')
```

```python
print(merged_gdf.geometry.head())
```

```python
merged_gdf.crs = 'EPSG:4326'
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
merged_gdf.plot(column='med_hh_income', ax=ax)
plt.title('Median Household Income Map')
plt.show()
```

```python
merged_gdf['ami_category'] = pd.Categorical(merged_gdf['ami_category'], categories=["Very Low Income", "Low Income", "Moderate Income", "Above Moderate Income"], ordered=True)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
merged_gdf.plot(column='below_med_income', ax=ax, legend=True, cmap='coolwarm')
ax.axis('off')
ax.set_title('Below Median Income by Census Tract', fontdict={'fontsize': '15', 'fontweight' : '3'})
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
merged_gdf.plot(column='below_60pct_med_income', ax=ax, legend=True, cmap='Reds')
ax.axis('off')
ax.set_title('Below 60% Median Income by Census Tract', fontdict={'fontsize': '15', 'fontweight' : '3'})
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
merged_gdf.plot(column='med_hh_income', ax=ax, legend=True, scheme='quantiles', k=5, cmap='YlGnBu')
ax.axis('off')
ax.set_title('Median Household Income by Census Tract - Quantiles', fontdict={'fontsize': '15', 'fontweight' : '3'})
plt.show()

```
